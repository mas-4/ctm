```python
with open('voting_record_dump109.txt', 'rt') as fin:
    records = list(fin)
records
records = [el.split() for el in records]

def transform_list_records(ls):
    return {l[0]: [int(i) for i in l[3:]] for l in ls}
records = transform_list_records(records)
records
```

Suppose $u$ and $v$ are two vectors. Let's take the simple case (relevant to the current lab) in which the entries are all 1, 0, or -1. Recall that the dot-product of $u$ and $v$ is defined as

$$
u \cdot v = \sum_k u[k]v[k]
$$

Consider the $k^{th}$ entry. If both $u[k]$ and $v[k]$ are 1, the corresponding term in the sum is 1.

```python
def policy_compare(sen_a, sen_b, votes):
    return sum([a*b for a,b in zip(votes[sen_a], votes[sen_b])])

policy_compare('Obama', 'McCain', records)
```

```python
def most_similar(sen, votes):
    others = list(votes.keys())
    others.pop(others.index(sen))
    maxv = 0
    maxs = None
    for o in others:
        if (v := policy_compare(sen, o, votes)) > maxv:
            maxv = v
            maxs = o
    return maxs, maxv
most_similar('Obama', records)
```

```python
def least_similar(sen, votes):
    others = list(votes.keys())
    others.pop(others.index(sen))
    minv = 10000
    mins = None
    for o in others:
        if (v := policy_compare(sen, o, votes)) < minv:
            minv = v
            mins = o
    return mins, minv
least_similar('McCain', records)
```
